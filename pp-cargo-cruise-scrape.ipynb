```python
#requesting url of website that contains all of the meetings
base_url = "https://meetings.boardbook.org/Public/Organization/1448"
response1 = requests.get(base_url)
```

```python
#isolating meeting links from page with all of the meetings

meetings = soup1.find_all("td", class_ = "nowrap")
meetings
```

```python
meeting_links = []

for a_tag in meetings:
    a_tags = a_tag.find("a")
    meeting_links.append(a_tags.get("href"))

meeting_links
    
#got meeting links tails 
```

```python
quarterly_report_links = []
link_count = 1

for each in attachment_details:
    print(f"Checking for quarterly report {link_count}/{len(attachment_details)}")
    link_count +=1
    snoozer = randrange(3,7)
    if "Cargo and Cruise Quarterly Report" in each:
        link = each.find("a").get("href")
        quarterly_report_links.append(link)
    else:
        print(f"No quarterly report")
    print(f"Snoozing for {snoozer}")
    time.sleep(snoozer)
        
#I stopped it because this was not working. my string filter failed.
```

```python
quarterly_reports_links = []

for each in attachment_details:
    link = each.find("a").get("href")
    name = each.find("span", class_="fileNameValue")
    if "Cargo and Cruise Quarterly Report" in name:
        link.replace("'","")
        quarterly_reports_links.append(link)

quarterly_reports_links

#the string filter worked. These are all of the links to the pages holding the quarterly reports.
```
